```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# df = pd.read_csv("heights.csv")
# df.head()
X = 2 * np.random.rand(100,1)  #rand: 난수 생성 함수
y = 4 +  3 * X + np.random.randn(100,1) #randn은 정규분포에 대한 난수 생성
```

```python
# X
```

```python
plt.plot(X, y, 'o')
plt.show()
```

```python
line_fitter = LinearRegression()
line_fitter.fit(X, y)
# X는 2차원 array형태 필요
```

```python
line_fitter.predict([[2.5]])
```

```python
# 기울기 : y증가량 / x증가량
line_fitter.coef_
```

```python
# 절편 (y절편)
line_fitter.intercept_
```

```python
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X))
plt.show()
```
